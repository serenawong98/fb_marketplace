```python
import clean_data
import os
import pandas as pd
```

```python
fb_clean = clean_data.CleanTabular(os.path.join(os.path.dirname(os.getcwd()), 'data/Products.csv'))
clean_series_text = ["product_name", 'product_description']
for i in clean_series_text:
    fb_clean.clean_text(i)
series_to_keep = ["product_name", 'product_description', 'location', 'price']
fb_clean.slice_df(series_to_keep)
fb_clean.clean_price_to_float('price')
fb_clean.clean_to_county('location')
fb_clean.clean_to_category_type('location')
```

```python
fb_clean.df.head(5)
```

```python
fb_clean.df.info()
```

```python
from sklearn import model_selection
```

```python
y = fb_clean.df['price']
x = fb_clean.df.drop(columns='price')
x.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
vectorizer = CountVectorizer(max_df=0.75, min_df = 0.01)
tfidf = TfidfTransformer()
x['product_description'] = x['product_description'].apply(lambda x: ' '.join(x))
vectorized_prod_des = vectorizer.fit_transform(x['product_description'])
tfidf_prod_des = tfidf.fit_transform(vectorized_prod_des)
features = tfidf.get_feature_names_out(vectorizer.get_feature_names_out())
df_prod_des = pd.DataFrame(tfidf_prod_des.toarray(), columns=features)
df_prod_des.head(5)
```

```python
x['product_name'] = x['product_name'].apply(lambda x: ' '.join(x))
vectorized_prod_name = vectorizer.fit_transform(x['product_name'])
tfidf_prod_name = tfidf.fit_transform(vectorized_prod_name)
features = tfidf.get_feature_names_out(vectorizer.get_feature_names_out())
df_prod_name = pd.DataFrame(tfidf_prod_name.toarray(), columns=features)
df_prod_name.head(5)
```

```python
x = x.drop(columns=["product_description", "product_name"])
x = pd.concat([x, df_prod_des, df_prod_name], axis = 1)
x.head(5)
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=0.3)
x_validation, x_test, y_validation, y_test = model_selection.train_test_split(x_test, y_test, test_size=0.5)
```

```python
from sklearn import linear_model
from sklearn import metrics
```

```python
model = linear_model.LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_train)
metrics.mean_squared_error(y_train, y_pred)
```

```python
y_test_pred = model.predict(x_test)
metrics.mean_squared_error(y_test_pred, y_test)
```

```python
y_validate_pred = model.predict(x_validation)
metrics.mean_squared_error(y_validate_pred, y_validation)
```
